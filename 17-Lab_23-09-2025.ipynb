```python
# Imports: Pandas and Matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Dataset and Preview Head
DataFrameDiabetes = pd.read_csv("./diabetes.csv")
print(DataFrameDiabetes.head())
```

```python
# Scatter Line Plot: Glucose vs Age
Figure, Axis = plt.subplots()
Axis.plot(DataFrameDiabetes["Age"], DataFrameDiabetes["Glucose"], color="green", marker="o", linestyle="-", label="Glucose vs Age")
Axis.set_xlabel("Age (Years)")
Axis.set_ylabel("Glucose Level (mg/dL)")
Axis.set_title("Diabetes Dataset: Glucose Level vs Age")
Axis.legend()
Axis.grid(True)
plt.show()
```

```python
AverageGlucoseByOutcome = DataFrameDiabetes.groupby('Outcome')['Glucose'].mean()
OutcomeCategories = ['Non-Diabetic' if x == 0 else 'Diabetic' for x in AverageGlucoseByOutcome.index]
AverageGlucose = AverageGlucoseByOutcome.values
Figure, Axis = plt.subplots(figsize=(8, 6))
Axis.bar(OutcomeCategories, AverageGlucose, color=['blue', 'orange'])
Axis.set_title('Average Glucose Level by Diabetes Outcome', fontsize=16, pad=20)
Axis.set_xlabel('Diabetes Outcome', fontsize=13)
Axis.set_ylabel('Average Glucose (mg/dL)', fontsize=13)
Axis.tick_params(axis='x', labelsize=12)
Axis.tick_params(axis='y', labelsize=12)

for IndexValue, GlucoseValue in enumerate(AverageGlucose):
    Axis.text(IndexValue, GlucoseValue + 2, f'{GlucoseValue:.1f}', ha='center', va='bottom', fontsize=11, fontweight='bold')

Axis.grid(axis='y', linestyle='--', alpha=0.7)
Axis.spines['top'].set_visible(False)
Axis.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```

# CHANGING IMAGE CHARACTERISTICS USING MATPLOTLIB

```python
from PIL import Image # Importing PIL library for image data

import matplotlib.pyplot as plt
import numpy as np
```

```python
ImageArray = np.asarray(Image.open("./dragonfly.jpeg"))  # Import image into Numpy array
print(repr(ImageArray))  # Each inner list represents a pixel (R,G,B)
```

```python
ImagePlot = plt.imshow(ImageArray)  # Plotting numpy arrays as images
```

```python
LuminosityImage = ImageArray[:, :, 0]  # Pseudocolor is only relevant to single-channel images
plt.imshow(LuminosityImage)  # Pick one channel (R) using slicing

# Add a colorbar to show the mapping of pixel intensity to color
plt.colorbar(label='Pixel Intensity')
plt.title('Pseudocolor of the Given Grayscale Image')
```

```python
# Print matrix shapes
print("Original image shape:", ImageArray.shape)   # (height, width, channels)
print("Luminosity image shape:", LuminosityImage.shape)   # (height, width)
print("\nOriginal pixel (0,0):", ImageArray[0, 0])      # full RGB values
print("Luminosity pixel (0,0):", LuminosityImage[0, 0])  # only Red channel value
```

```python
# Print repr of both
print("repr(ImageArray) [truncated]:", repr(ImageArray)[:300], "...\n")
print("repr(LuminosityImage) [truncated]:", repr(LuminosityImage)[:300], "...\n")
```

```python
# Expand LuminosityImage to 3 channels for fair diff
LuminosityExpanded = np.stack([LuminosityImage, LuminosityImage, LuminosityImage], axis=-1)
DifferenceMatrix = ImageArray - LuminosityExpanded
print("\nDifference matrix sample (5x5 pixels):")
print(DifferenceMatrix[:5, :5, :])
```

```python
plt.imshow(LuminosityImage, cmap="hot")
```

```python
# List of colormaps to demonstrate
ColorMaps = ["gray", "viridis", "plasma", "inferno", "magma", "cividis", "hot", "cool", "spring", "summer", "autumn", "winter"]
# Plot all colormaps
Figure, Axes = plt.subplots(3, 4, figsize=(12, 9))
for AxisLocal, ColorMap in zip(Axes.ravel(), ColorMaps):
    AxisLocal.imshow(LuminosityImage, cmap=ColorMap)
    AxisLocal.set_title(ColorMap)
    AxisLocal.axis("off")

plt.tight_layout()
plt.show()
```

# INVERSE OR NEGATIVE OF THE IMAGE

```python
ImageObject = Image.open("./dragonfly.jpeg")  # Load image
ImageObject = ImageObject.convert("RGB")  # Ensure 3 channels
ImageArray = np.array(ImageObject)
```

```python
print("Original Image Array:", ImageArray)
```

```python
# Find the inverse (negative image)
# For 8-bit images, max pixel value is 255
NegativeImageArray = 255 - ImageArray
print("Inverted Image Array", NegativeImageArray)
```

```python
# Display original and negative side by side
Figure, Axes = plt.subplots(1, 2, figsize=(10, 5))
Axes[0].imshow(ImageArray)
Axes[0].set_title("Original Image")
Axes[0].axis("off")
Axes[1].imshow(NegativeImageArray)
Axes[1].set_title("Negative (Inverted) Image")
Axes[1].axis("off")
plt.show()
```

# CATEGORICAL DATA

```python
# Plot count of diabetic vs non-diabetic patients
plt.figure(figsize=(6, 4))
SeriesCounts = DataFrameDiabetes['Outcome'].replace({0: "Non-Diabetic", 1: "Diabetic"}).value_counts()
SeriesCounts.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title("Count of Diabetes Outcome")
plt.xlabel("Outcome")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
