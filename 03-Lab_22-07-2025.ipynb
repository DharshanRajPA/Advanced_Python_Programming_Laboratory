```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

visited_locations = input("Enter places you have already traveled : ").split(",")             
visited_locations = [place.strip().title() for place in visited_locations]

user_scores = []                                                                                
for place in visited_locations:
    rate = int(input(f"Give your rating for {place} (1-5): "))
    user_scores.append(rate)

print("\nYour Travel History and Ratings:")                                                 
for i in range(len(visited_locations)):
    print(f"{visited_locations[i]}: {user_scores[i]} stars")

recommendation_list = []

for i in range(len(user_scores)):                                                               
    if user_scores[i] >= 4:
        recommendation_list.append(f"{visited_locations[i]}")

print("\nInitial Recommendation List:", recommendation_list)

recommendation_list.append("Tokyo")                                                         
print("After append:", recommendation_list)

recommendation_list.insert(1, "London")                                                    
print("After insert at index 1:", recommendation_list)

removed = recommendation_list.pop()                                                          
print(f"After pop (removed {removed}):", recommendation_list)

if "London" in recommendation_list:
    recommendation_list.remove("London")
    print("After removing 'London':", recommendation_list)

recommendation_list.sort()                                                                   
print("After sort:", recommendation_list)

if len(recommendation_list) > 1:
    del recommendation_list[1]
    print("After delete at index 1:", recommendation_list)

other_travelers = {                                                                           
    "Traveler1": {"Tokyo": 5, "Venice": 4, "Berlin": 2},
    "Traveler2": {"London": 3, "Tokyo": 4, "Mumbai": 5},
    "Traveler3": {"Venice": 5, "Berlin": 3, "London": 4}
}

all_places = list(set(visited_locations + [place for user in other_travelers.values() for place in user]))    

def create_vector(user_dict):
    return [user_dict.get(place, 0) for place in all_places]

current_user_vector = create_vector(dict(zip(visited_locations, user_scores)))

print("\nCollaborative Filtering Recommendations:")                                                       
for user, user_rating in other_travelers.items():
    other_vector = create_vector(user_rating)
    sim = cosine_similarity([current_user_vector], [other_vector])[0][0]
    print(f"Similarity with {user}: {sim:.2f}")
    if sim > 0.5:
        for place in user_rating:
            if place not in visited_locations and place not in recommendation_list:
                recommendation_list.append(place)

print("\nFinal Recommendation List After Collaborative Filtering:", recommendation_list)
```
