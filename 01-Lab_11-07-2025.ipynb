```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Setting the style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Creating a basic line chart for data visualization
x_values = [2, 4, 6, 8]
y_values = [8, 14, 4, 10]

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, linewidth=3, marker='o', markersize=8)
plt.title('Data Analysis Chart', fontsize=16, fontweight='bold')
plt.xlabel('Independent Variable', fontsize=12)
plt.ylabel('Dependent Variable', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Comprehensive data visualization with multiple elements
x_data = [2, 4, 6, 8]
y_data = [8, 12, 10, 18]

plt.figure(figsize=(12, 8))

# Primary line plot
plt.plot(x_data, y_data, linewidth=3, color='navy', 
         marker='o', markersize=12, label='Primary Dataset', alpha=0.8)

# Scatter plot for data points
plt.scatter(x_data, y_data, s=150, color='crimson', 
           alpha=0.9, label='Data Points', edgecolors='black', linewidth=2)

# Adding trend line using seaborn
sns.regplot(x=x_data, y=y_data, scatter=False, 
           line_kws={'color': 'green', 'linestyle': ':', 'linewidth': 2})

plt.title('Multi-Element Data Visualization', fontsize=20, fontweight='bold')
plt.xlabel('X Axis Description', fontsize=14)
plt.ylabel('Y Axis Description', fontsize=14)
plt.legend(fontsize=12, loc='upper left')
plt.grid(True, alpha=0.3)
plt.xlim(0, 12)
plt.ylim(0, 25)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Creating sample data for advanced visualizations
np.random.seed(42)
data_points = 50
x_vals = np.random.normal(0, 1, data_points)
y_vals = np.random.normal(0, 1, data_points)
categories = np.random.choice(['A', 'B', 'C'], data_points)

# Creating a DataFrame for seaborn
df = pd.DataFrame({
    'X_Coordinate': x_vals,
    'Y_Coordinate': y_vals,
    'Category': categories
})

# Advanced visualization with seaborn
plt.figure(figsize=(14, 10))

# Creating subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# 1. Scatter plot with hue
sns.scatterplot(data=df, x='X_Coordinate', y='Y_Coordinate', 
                hue='Category', ax=ax1, s=100, alpha=0.7)
ax1.set_title('Scatter Plot with Categories', fontweight='bold')
ax1.grid(True, alpha=0.3)

# 2. Distribution plot
sns.histplot(data=df, x='X_Coordinate', bins=15, ax=ax2, alpha=0.7)
ax2.set_title('Distribution of X Values', fontweight='bold')
ax2.grid(True, alpha=0.3)

# 3. Box plot
sns.boxplot(data=df, x='Category', y='Y_Coordinate', ax=ax3)
ax3.set_title('Box Plot by Category', fontweight='bold')
ax3.grid(True, alpha=0.3)

# 4. Violin plot
sns.violinplot(data=df, x='Category', y='X_Coordinate', ax=ax4)
ax4.set_title('Violin Plot by Category', fontweight='bold')
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Creating a comprehensive dashboard
plt.style.use('default')
sns.set_theme(style="whitegrid")

# Sample data for different chart types
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [23, 45, 56, 78]
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']

# Creating a figure with multiple subplots
fig = plt.figure(figsize=(16, 12))

# 1. Bar chart
ax1 = plt.subplot(2, 3, 1)
bars = plt.bar(categories, values, color=colors, alpha=0.8, edgecolor='black')
plt.title('Bar Chart Analysis', fontweight='bold', fontsize=14)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(rotation=45)

# 2. Pie chart
ax2 = plt.subplot(2, 3, 2)
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Pie Chart Distribution', fontweight='bold', fontsize=14)

# 3. Line chart
ax3 = plt.subplot(2, 3, 3)
x_line = np.arange(len(categories))
plt.plot(x_line, values, marker='o', linewidth=3, markersize=8, color='navy')
plt.title('Line Chart Trend', fontweight='bold', fontsize=14)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(x_line, categories, rotation=45)

# 4. Area chart
ax4 = plt.subplot(2, 3, 4)
plt.fill_between(categories, values, alpha=0.6, color='lightblue')
plt.plot(categories, values, marker='s', linewidth=2, color='darkblue')
plt.title('Area Chart', fontweight='bold', fontsize=14)
plt.xticks(rotation=45)

# 5. Horizontal bar chart
ax5 = plt.subplot(2, 3, 5)
plt.barh(categories, values, color=colors, alpha=0.8)
plt.title('Horizontal Bar Chart', fontweight='bold', fontsize=14)
plt.xlabel('Values')

# 6. Scatter plot
ax6 = plt.subplot(2, 3, 6)
x_scatter = np.random.normal(0, 1, 100)
y_scatter = np.random.normal(0, 1, 100)
plt.scatter(x_scatter, y_scatter, alpha=0.6, color='crimson')
plt.title('Scatter Plot', fontweight='bold', fontsize=14)
plt.xlabel('X Values')
plt.ylabel('Y Values')

plt.tight_layout()
plt.show()
```
