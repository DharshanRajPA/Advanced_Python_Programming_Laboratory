### Program 1: Sentiment Analysis Using TextBlob


```python
from textblob import TextBlob                    # Import TextBlob Library For Sentiment Analysis

UserInputText = input("Enter a sentence: ")      # Get User Input Text From Console
TextBlobObject = TextBlob(UserInputText)         # Create TextBlob Object From Input Text

SentimentPolarity = TextBlobObject.sentiment.polarity      
# Extract Sentiment Polarity Value (-1 to 1)
SentimentSubjectivity = TextBlobObject.sentiment.subjectivity  
# Extract Sentiment Subjectivity Value (0 to 1)

print("Polarity:", SentimentPolarity)            # Display Polarity Score
print("Subjectivity:", SentimentSubjectivity)    # Display Subjectivity Score

if SentimentPolarity > 0:                        # Check If Polarity Is Positive
    print("Sentiment: Positive")                 # Display Positive Sentiment
elif SentimentPolarity < 0:                      # Check If Polarity Is Negative
    print("Sentiment: Negative")                  # Display Negative Sentiment
else:                                            # If Polarity Is Zero
    print("Sentiment: Neutral")                  # Display Neutral Sentiment
```

### Program 4: Weather Information Retrieval Using Requests Library


```python
import requests                                  # Import Requests Library For HTTP Requests
import json                                      # Import JSON Library For Data Parsing

CityName = input("Enter city name: ")           # Get City Name From User Input
WeatherAPIKey = "your_api_key_here"              # Placeholder For Weather API Key
WeatherAPIURL = f"http://api.openweathermap.org/data/2.5/weather?q={CityName}&appid={WeatherAPIKey}&units=metric"  
# Construct Weather API URL With City Name And API Key

try:                                            # Try Block For Error Handling
    WeatherResponse = requests.get(WeatherAPIURL)  
    # Send HTTP GET Request To Weather API
    
    if WeatherResponse.status_code == 200:       # Check If Request Was Successful
        WeatherData = WeatherResponse.json()      # Parse JSON Response Data
        
        Temperature = WeatherData['main']['temp']  
        # Extract Temperature From Response
        Humidity = WeatherData['main']['humidity']  
        # Extract Humidity From Response
        Description = WeatherData['weather'][0]['description']  
        # Extract Weather Description From Response
        
        print(f"Weather Information For {CityName}:")  
        # Display City Name Header
        print(f"Temperature: {Temperature}Â°C")    # Display Temperature
        print(f"Humidity: {Humidity}%")           # Display Humidity
        print(f"Description: {Description}")      # Display Weather Description
    else:                                        # If Request Failed
        print("Failed To Retrieve Weather Data")  
        # Display Error Message
        print("Please Check Your API Key And City Name")  
        # Display Helpful Message
except Exception as ErrorMessage:               # Catch Any Exceptions
    print(f"Error Occurred: {ErrorMessage}")     # Display Error Message
```

### Program 5: NumPy Array Manipulations - Basic Operations


```python
import numpy as np                               # Import NumPy Library As np

# Creating NumPy Array From List
OriginalList = [1, 2, 3, 4, 5, 6]               # Define Original List Of Numbers
NumpyArray = np.array(OriginalList)              # Convert List To NumPy Array
print("Original NumPy Array:", NumpyArray)      # Display Original Array

# Reshaping Array
ReshapedArray = NumpyArray.reshape(2, 3)        # Reshape Array To 2x3 Matrix
print("\nReshaped Array (2x3):\n", ReshapedArray)  
# Display Reshaped Array

# Array Slicing
SlicedArray = NumpyArray[1:4]                   # Extract Elements From Index 1 To 3
print("\nSliced Array (Index 1-3):", SlicedArray)  
# Display Sliced Array

# Mathematical Operations
ArraySum = np.sum(NumpyArray)                    # Calculate Sum Of All Elements
ArrayMean = np.mean(NumpyArray)                  # Calculate Mean Of All Elements
ArrayMax = np.max(NumpyArray)                    # Find Maximum Value In Array
ArrayMin = np.min(NumpyArray)                    # Find Minimum Value In Array

print("\nArray Statistics:")                     # Display Statistics Header
print(f"Sum: {ArraySum}")                        # Display Sum
print(f"Mean: {ArrayMean}")                      # Display Mean
print(f"Maximum: {ArrayMax}")                    # Display Maximum
print(f"Minimum: {ArrayMin}")                    # Display Minimum

# Array Operations
ArraySquare = NumpyArray ** 2                    # Square Each Element
print("\nSquared Array:", ArraySquare)            # Display Squared Array

ArrayDouble = NumpyArray * 2                     # Multiply Each Element By 2
print("Doubled Array:", ArrayDouble)             # Display Doubled Array
```

### Program 6: Threading Example - File Download Using Multiple Threads


```python
import threading                                  # Import Threading Module
import requests                                    # Import Requests Library
from pathlib import Path                          # Import Path From pathlib Module

# Function To Download File From Given URL And Save Locally
def DownloadFile(DownloadURL, FileName):           # Define Download Function With Parameters
    print(f'Downloading {DownloadURL} to {FileName}')  
    # Display Download Start Message
    
    FileResponse = requests.get(DownloadURL)       # Send HTTP GET Request To Fetch File Content
    Path(FileName).write_bytes(FileResponse.content)  
    # Save File Content To Local Storage
    
    print(f'Finished Downloading {FileName}')     # Display Download Complete Message

# Base URL Should Point To Raw GitHub Content, Not The Repo Page
# Example Format: https://raw.githubusercontent.com/<username>/<repo>/<branch>/<path-to-file>
BaseURL = "https://github.com/DharshanRajPA/Advanced_Python_Programming_Laboratory/blob/main/17-Lab_23-09-2025.ipynb"  
# Define Base URL For File Downloads

# List Of Files To Download From GitHub Repository
DownloadURLs = [                                  # Define List Of URLs
    f'{BaseURL}',                                 # First File URL
    f'{BaseURL}',                                 # Second File URL
]

ThreadObjects = []                                # Initialize List To Store Thread Objects

# Loop Through URLs And Start Downloading Files Concurrently
for URL in DownloadURLs:                          # Iterate Through Each URL
    FileName = "Downloads/" + URL.split("/")[-1]  # Generate File Name From URL
    ThreadInstance = threading.Thread(target=DownloadFile, args=(URL, FileName))  
    # Create New Thread For File Download
    
    ThreadInstance.start()                        # Start Thread Execution
    ThreadObjects.append(ThreadInstance)          # Add Thread To Tracking List

# Ensure All Downloads Finish Before Moving Ahead
[ThreadInstance.join() for ThreadInstance in ThreadObjects]  
# Wait For All Threads To Complete
```
